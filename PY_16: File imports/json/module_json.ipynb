```python
import json
from pprint import pprint
```

→ Чтобы перевести данные из формата JSON в формат, который можно обрабатывать инструментами Python, необходимо выполнить процедуру, которая называется **десериализация (декодирование данных)**. Обратный процесс, связанный с переводом структур данных Python в формат JSON, называется **сериализацией**.

```python
with open('recipes.json') as f:
    recipes = json.load(f)
pprint(recipes)
```

```python
print(len(recipes[0]['ingredients']))
```

```python
for rec in recipes:
    if rec['id'] == 13121:
        print(rec['cuisine']) 
```

```python
unique_cuisines = set()
unique_cuisines_dct = dict()
for rec in recipes:
    cuisine = rec['cuisine']
    unique_cuisines.add(cuisine)
    if cuisine in unique_cuisines_dct: 
        unique_cuisines_dct[cuisine] += 1
    else:
        unique_cuisines_dct[cuisine] = 1
    
    
print("Total cuisines:", len(unique_cuisines))
print("Most popular cuisine:", max(unique_cuisines_dct, key=unique_cuisines_dct.get))
```

```python
# json -> pandas DF
import pandas as pd

df = pd.DataFrame(recipes)
df.head()
```

```python
ingredients = set()
for i, row in df.iterrows():
    ingredients.update(row['ingredients'])

print('Number of unique ingredients:', len(ingredients))

for ingredient in list(ingredients):
    df[ingredient] = df['ingredients'].apply(lambda x: 1 if ingredient in x else 0)

df.head()
```

```python
df.to_csv('recipes.csv', index=False)
```

#### Re-creating the json structure out of the DF

```python
import json

def make_list(row):
    ingredients = []
    filtered_df = row.iloc[:, 3:]
    for column in filtered_df:
        if int(filtered_df[column]) == 1:
            ingredients.append(column)
    return ingredients

ids = []
for i, row in df.iterrows():
    id = row['id']
    if id not in ids:
        ids.append(id)
        
ingredients = df.columns.tolist()[3:]

new_recipes = []
for current_id in ids:
    cuisine = df[df['id'] == current_id]['cuisine'].iloc[0]
    current_ingredients = make_list(df[df['id'] == current_id])
    current_recipe = {'cuisine': cuisine, 'id': int(current_id), 'ingredients': current_ingredients}
    new_recipes.append(current_recipe)
    
# exporting to json
new_recipes = json.dumps(new_recipes)
with open('../data/new_recipes.json', 'w') as write_file:
    write_file.write(new_recipes)
```
